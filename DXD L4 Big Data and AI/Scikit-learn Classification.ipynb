# Classification

* Using values of X predict a category Y
* Breast cancer dataset is popular to classify Malignant and Benign tumour

```python
from sklearn.datasets import load_breast_cancer

bc = load_breast_cancer()
print(dir(bc))
print(bc.DESCR)
```

```python
df_breast = pd.DataFrame(bc.data, columns=bc.feature_names)
df_breast.head()
```

```python
df_breast.isna().sum()
```

```python
df_breast['type'] = bc.target 
```

```python
df_breast.head()
```

```python
df_breast['type'].value_counts()
# 0 - Malignant
# 1 - Benign
```

```python
bar = df_breast['type'].value_counts().plot(kind='bar')
bar.set_xticklabels(['B','M'])
```

# Modelling

* We will just use Worst Perimeter points
    * In real life we would be using multiple variables
* We will split our data into test and training sets
    * 75% train, 25% test
* Use logistic regression to predict the type of tumour
* `.fit()' will train our model - LEARNING

```python
x = df_breast['worst perimeter']
x = x[:, np.newaxis]
y = df_breast['type']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 2)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
val = np.linspace(20, 300, 1000).reshape(-1,1)
train_pred_prob = model.predict_proba(val) 
train_pred_prob #probability of 0, probability of 1
```

```python
train_pred = model.predict(val) #Predict group
```

We will visualise the probality of 1 (Benign) 

```python
#Get probability curve of our training values
sns.scatterplot(x_train.ravel(), y_train)
sns.lineplot(val.ravel(), train_pred_prob[:,1].ravel())
sns.lineplot(val.ravel(), train_pred.ravel())
```

# predict

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_test.value_counts()
```

```python
sns.scatterplot(x_test.ravel(), y_pred, label='pred')
```

```python
sns.scatterplot(x_test.ravel(), y_test, label='actual')
```

# Evaluate

Confusion matrix and classification reort

* **Precision** = $\frac{TP}{TP+FP}$ - From all that we predicted as positive how many were actually positive


* **Recall** = $\frac{TP}{TP+FN}$ - From what was actually positive how many did we correcly predict as positive

* **F1-score** - average between precision and recall

```python
con = pd.DataFrame({
    'Actually Positive' : ['TP', 'FN'],
    'Actually Negative' : ['FP', 'TN']
})

con.index = ['Predicted Positive', 'Predicted Negative']
con
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred,
                      target_names=['Malignant', 'Benign']))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test, y_pred,
                      labels=[0, 1]))
```
