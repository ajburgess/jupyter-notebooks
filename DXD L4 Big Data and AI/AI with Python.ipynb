# Key Modules for AI with Python

* There are many libraries available for AI. Below is a list of more commonly used modules, and an explanation of what they do and why they are used.

# NumPy

* Numerical Python
* More efficient than Python with large amounts of data
* Supports multi-dimensional arrays and matrices which allow a collection of high-level mathematical function to be performed on them

# Pandas

* Python Data Analysis Library
* Large datasets/ high-level data structures
* Data manipulation
    * Merging and join tables
    * Filter data
* Time series data
* Uses Numpy arrays

# Matplotlib

* 2D plotting library 
* Basis of other plotting packages

# Seaborn

* Data visualisation tool
* Resembles MATLAB plots

# SciPy

* Uses Numpy for more mathematical functions
* Modules for common tasks in scientific programmaing
    * Differential equations
    * Intergreations
    * Linear algerba, etc.

# Scikit-learn

* Machine learning library
* Regression, clustering amd classification algorithms available
* Supports NumPy and Scipy

# TensorFlow

* Create deep learning models
* Created and released by Google

# Keras

* Neural Network library
* Can run ontop of TensorFlow, Theano and others

# PyTorch

* Machine learning library
* Computer vision and Natural Language Procressing (NLP) tasks
* Developed by Facebooks AI research group

# Theano

* Manipulate and evaluate mathematical expressions in particular matrix-values one
* Use NumPy syntax
